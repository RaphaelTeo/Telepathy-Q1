```python
import pandas as pd
```

```python
df_canon = pd.read_excel('canonical models.xlsx')
df_m1 = pd.read_excel('car models 1.xlsx')
df_m2 = pd.read_excel('car models 2.xlsx')
```

```python
# Canonical model- data cleaning and create keys
df_canon_renamed = df_canon.rename({'make':'canonical_make', ' model':'canonical_model'}, axis=1)
df_canon_keyed = df_canon_renamed.copy()
df_canon_keyed['canonical_make_key'] = df_canon_keyed['canonical_make'].apply(lambda x: x.lower().strip())
df_canon_keyed['canonical_model_key'] = df_canon_keyed['canonical_model'].apply(lambda x: x.lower().strip())
df_canon_keyed['join_key'] = df_canon_keyed['canonical_make_key'] + df_canon_keyed['canonical_model_key']
df_canon_keyed
```

```python
# Model 1- data cleaning and create keys
df_m1_renamed = df_m1.rename({'Make':'make', 'Model':'model', 'MakeDescription': 'make_desc', 'ModelDescription':'model_desc'}, axis=1)
df_m1_keyed = df_m1_renamed.copy()
df_m1_keyed['make_desc_key'] = df_m1_keyed['make_desc'].apply(lambda x: x.lower().strip())
df_m1_keyed['model_desc_key'] = df_m1_keyed['model_desc'].apply(lambda x: x.lower().strip())
df_m1_keyed['join_key'] = df_m1_keyed['make_desc_key'] + df_m1_keyed['model_desc_key']
df_m1_keyed
```

```python
# Augment Canonical model with Model 1
df_m1_merged = df_canon_keyed.merge(df_m1_keyed, how='left', on = 'join_key', indicator=True)
df_m1_merged = df_m1_merged.rename({'_merge':'_merge 1'}, axis=1) # rename indicator column
df_m1_merged
```

```python
def fill_in(df, data_complete, data_incomplete):

    print(f'{data_complete} vs {data_incomplete}')
    
    # for each unique entry in the data_complete column 
    for unique_value in pd.unique(df[data_complete]):
        
        # see the various options (or NA) in the data_incomplete column
        grouped = df[df[data_complete] == unique_value].groupby([data_incomplete], sort=True, dropna=True).groups # sort puts the entry with highest count first. groups returns a dict of the unique values found and its indices
        print(grouped)

        # if there is no reference data, just use the unique value from the column with complete data
        if len(grouped) == 0:
            correct_make_name = unique_value

        else:
            # count the number of occurrences of each option that appeared in the data_incomplete column
            for each_make in grouped:
                grouped[each_make] = len(grouped[each_make])
            print(grouped)

            # assume that the option which appears the most is the correct option
            correct_make_name = max(grouped, key=grouped.get) 
        
        # update the data_incomplete column for this unique entry with the correct option
        df.loc[df[data_complete] == unique_value, data_incomplete] = correct_make_name

    return df
```

```python
for data_complete, data_incomplete in zip(['canonical_make', 'canonical_model', 'canonical_make', 'canonical_model'], ['make', 'model', 'make_desc', 'model_desc']):
    df_m1_merged = fill_in(df_m1_merged, data_complete, data_incomplete)
```

```python
# Formatting to all caps
df_m1_merged['model'] = df_m1_merged['model'].apply(lambda x: x.upper())
df_m1_merged['model_desc'] = df_m1_merged['model_desc'].apply(lambda x: x.upper())

# Remove mapping keys
cols_to_drop = ['canonical_make_key', 'canonical_model_key', 'join_key', 'make_desc_key', 'model_desc_key', '_merge 1']
df_m1_merged.drop(columns=cols_to_drop, inplace=True)
df_m1_merged
```

```python
# Export to excel
df_m1_merged.to_excel('Submission.xlsx', index=False)
```
