```python
import pandas as pd
```

```python
df_canon = pd.read_excel('canonical models.xlsx')
df_m1 = pd.read_excel('car models 1.xlsx')
df_m2 = pd.read_excel('car models 2.xlsx')


```

```python
# Data cleaning and to create keys for canonical models
df_canon_renamed = df_canon.rename({'make':'canonical_make', ' model':'canonical_model'}, axis=1)
df_canon_keyed = df_canon_renamed.copy()
df_canon_keyed['canonical_make_key'] = df_canon_keyed['canonical_make'].apply(lambda x: x.lower().strip())
df_canon_keyed['canonical_model_key'] = df_canon_keyed['canonical_model'].apply(lambda x: x.lower().strip())
df_canon_keyed['join_key'] = df_canon_keyed['canonical_make_key'] + df_canon_keyed['canonical_model_key']
df_canon_keyed
```

```python
# Data cleaning and to create keys for models 1
df_m1_renamed = df_m1.rename({'Make':'make', 'Model':'model', 'MakeDescription': 'make_desc', 'ModelDescription':'model_desc'}, axis=1)
df_m1_keyed = df_m1_renamed.copy()
df_m1_keyed['make_desc_key'] = df_m1_keyed['make_desc'].apply(lambda x: x.lower().strip())
df_m1_keyed['model_desc_key'] = df_m1_keyed['model_desc'].apply(lambda x: x.lower().strip())
df_m1_keyed['join_key'] = df_m1_keyed['make_desc_key'] + df_m1_keyed['model_desc_key']
df_m1_keyed
```

```python
# Do merging based on the keys
df_m1_merged = df_m1_keyed.merge(df_canon_keyed, how='left', on = 'join_key', indicator=True)
df_m1_merged
```

```python
# Do merging based on the keys
df_m1_merged = df_canon_keyed.merge(df_m1_keyed, how='right', on = 'join_key', indicator=True)
df_m1_merged
```

```python
# To see which are the entries that did not match canonical models
df_m1_merged[df_m1_merged['_merge'] == 'right_only']
```

```python
# Clean up for presentation
df_m1_merged.drop(columns=[''])
```
